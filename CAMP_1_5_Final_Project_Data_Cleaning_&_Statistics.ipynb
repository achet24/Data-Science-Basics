##Histogram
<font color="green" >This is a frequency count of a variable.
A histogram is the graphical representation of data where data is grouped into continuous number ranges and each range corresponds to a vertical bar.

<font color="green" >*Note: *bin value indicates the number of columns you want in your histogram.

<font color="green" >Smaller bin value = less detailed = fatter columns = more data in each "bar" of the histogram

<font color="green" >Bigger bin value = more detailed = skinnier columns = less data in each "bar" of the histogram

plt.figure()

Remember to think about what variables work best for each chart type.

* üìà Line plot -> this shows how numerical data changes over time. The y-axis should be a numerical variable (integer or float data type) and the x-axis should be a variable related to time (e.g., hours, days, months).
* üìâ Scatter plot -> this shows the relationship between two continuous variables (float data type).
* üìú Histogram -> this shows the frequency distribution of a numerical variable (integer or float data type).
* üìä Bar chart -> helps represent categorical data. The variable type used for this graph should be object/string. It may also be an integer data type in some circumstances. For example, there might be 3 categories for fire hazard severity: 1, 2, and 3.
